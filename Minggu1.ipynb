```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow 
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt 
```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]
```

```python
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0),fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image,image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
image_2 = cv.cvtColor(image,cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image,cv.COLOR_BGR2RGB)

for y in range(lebar):
  image_3[int((tinggi)/2),y] = [255,255,255] #sintax adalah koordinat untuk menarik garis awal dan garis akhir 

final_frame = cv.hconcat((image_2,image_3))
cv2_imshow(final_frame)
```

```python
for y in range(tinggi):
  image_3[y,int((lebar)/2)] = [0,0,0] 
final_frame = cv.hconcat((image_2,image_3))

cv2_imshow(final_frame)
```

```python
for y in range(tinggi):
  cv.line(image_3, (0,0), (lebar, tinggi), (0,0,0), 1) #menarik gari dengan titik awal dari pojok kiri atas ke pojok kanan bawah koordinat 2 adalah titik awal 3 titik akhir titik 4 adalah warna
  cv.line(image_3, (lebar,0), (0, tinggi), (0,0,0), 1) #dan sebaliknya
final_frame = cv.hconcat((image_2,image_3))            #Menggabungkan kedua gambar  
cv2_imshow(final_frame)
```
